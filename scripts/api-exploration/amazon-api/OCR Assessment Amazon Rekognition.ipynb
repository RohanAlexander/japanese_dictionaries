```python
#image_path = path.expanduser(r'E:\Users\Usuario\Documents\GitHub_Website\website\japanese_dictionaries\inputs\00-sample-pages_1-cropped.png')
image_path = path.expanduser(r'E:\Users\Usuario\Downloads\text_in_image.png')
```

```python
key_id = os.environ['aws_access_key_id']
secret_key = os.environ['aws_secret_access_key']
session = os.environ['aws_session_token']
client = boto3.client('rekognition', aws_access_key_id = key_id, 
                      aws_secret_access_key=secret_key, 
                      aws_session_token=session,
                      region_name='us-east-1')
```

```python
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read())
```

```python
encoded_image = encode_image(image_path)
```

```python
import io

img = Image.open(image_path, mode='r')
roiImg = img

imgByteArr = io.BytesIO()
roiImg.save(imgByteArr, format='PNG')
imgByteArr = imgByteArr.getvalue()
```

```python
response = client.detect_text(Image={
        'Bytes': imgByteArr})
```
